```python
import numpy as np
from scipy.misc import imread, imresize
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

# Read Image & Check Shape, type

```python
dog = imread("./img/dog.jpeg")

print(type(dog))
print(dog.shape)
```

# Show Image

```python
plt.figure(0)
plt.imshow(dog)
plt.title("DOG")
plt.draw()
```

# Resize

```python
dog_up = imresize(dog, [2000, 2000, 3])
plt.figure(0)
plt.imshow(dog_up)
plt.title("DOG_sizeup")
plt.draw()
```

# Grayscale

```python
def rgb2gray(image) :
    if len(image.shape) is 3 :
        return np.dot(image[...,:3], [0.299,0.584,0.114])
    else : 
        print("Check Image !")
        return image
```

```python
dog_gray = rgb2gray(dog)
plt.imshow(dog_gray, cmap = plt.get_cmap("gray"))
plt.colorbar()
plt.draw()
```

# Reshape

### Matrix -> Vector

```python
dog_vec = np.reshape(dog_gray, (1, -1))
print(dog_vec.shape)
```

### Vector -> Matrix

```python
dog_matrix = np.reshape(dog_vec, (1600, -1))
print(dog_matrix.shape)
```

# Read Images from Folder

```python
cwd = os.getcwd()
path = cwd + "/img" 
Img_list = os.listdir(path)
print(len(Img_list),"file are in", path)
```

```python
for i, name in enumerate(Img_list) :
    print(i,"file's name is", name)
```

```python
image_label = [".jpg", ".png", ".jpeg", ".gif"]
imgs = []
names = []
for f in Img_list :
    ext = os.path.splitext(f)[1]
    if ext.lower() not in image_label:
        continue
    # Append imgs
    fullpath = os.path.join(path,f)
    imgs.append(imread(fullpath))
    # Append names
    names.append(os.path.splitext(f)[0])
```

```python
for img, name in zip(imgs, names) :
    plt.imshow(img)
    plt.title(name)
    plt.show()
```
